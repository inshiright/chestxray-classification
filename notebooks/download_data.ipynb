```python
from kaggle.api.kaggle_api_extended import KaggleApi

# Load environment variables from the .env file
# Ensure the .env file is uploaded to the Colab workspace path (e.g., /content/.env)


# Define the destination path in Colab's instance memory
dataset_path = '/content/dataset'
os.makedirs(dataset_path, exist_ok=True)

print("Authenticating with Kaggle API...")
api = KaggleApi()
api.authenticate()

print(f"Downloading and extracting NIH dataset to {dataset_path}...")
api.dataset_download_files('nih-chest-xrays/data', path=dataset_path, unzip=True)

print("Download and extraction complete!")
```
